```python
import requests
import json
import pandas as pd
import time
from datetime import datetime

jobs = ["Software Engineer",
    "Senior Software Engineer",
    "Full Stack Developer",
    "Frontend Developer",
    "Backend Developer",
    "DevOps Engineer",
    "Cloud Solutions Architect",
    "Cloud Engineer",
    "Data Scientist",
    "Machine Learning Engineer",
    "AI Specialist",
    "Deep Learning Engineer",
    "Data Engineer",
    "Big Data Engineer",
    "Business Intelligence Developer",
    "Cybersecurity Engineer",
    "Information Security Analyst",
    "Security Architect",
    "Penetration Tester (Ethical Hacker)",
    "Network Architect",
    "Systems Architect",
    "Mobile Application Developer",
    "Software Architect",
    "Product Manager (Tech)",
    "Site Reliability Engineer (SRE)",
    "Database Administrator (DBA)",
    "Blockchain Developer",
    "IoT Developer",
    "AR/VR Developer",
    "Robotics Engineer"
]
len(jobs)
```

```python


import requests
import json

API_KEY = "YOUR_API_KEY"  # Get it from: https://jooble.org/api/about
URL = f"https://jooble.org/api/b8bf7c52-c723-49da-ab8a-71f6ab41765c"
RESULTS_PER_PAGE = 20   # Jooble max per page
MAX_RESULTS = 300       # max per job title
HEADERS = {"Content-Type": "application/json"}


all_jobs = []

for title in jobs :
    print(f"\n🔍 Searching: {title}")
    collected = 0
    page = 1

    while collected < MAX_RESULTS:
        needed = min(RESULTS_PER_PAGE, MAX_RESULTS - collected)

        payload = {
            "keywords": title,
            "location": "",          # empty = worldwide
            "page": page,
            "ResultOnPage": needed,
            "datePosted": 1          # 1 = last 30 days
        }

        try:
            response = requests.post(URL, headers=HEADERS, data=json.dumps(payload), timeout=15)

            if response.status_code != 200:
                print(f"  ❌ Error {response.status_code} on page {page}")
                break

            data = response.json()
            jobs_page = data.get("jobs", [])

            if not jobs_page:
                print(f"  ✅ No more results at page {page} ({collected} total)")
                break

            for job in jobs_page:
                all_jobs.append({
                    "Position":        job.get("title", "N/A").strip(),
                    "Company":         job.get("company", "N/A").strip(),
                    "Location":        job.get("location", "N/A").strip(),
                    "Job_Description": job.get("snippet", "N/A").strip(),
                    "URL":             job.get("link", "N/A").strip(),
                    "Search_Title":    title,                             # which title found this job
                    "Date_Scraped":    datetime.today().strftime("%Y-%m-%d")
                })

            collected += len(jobs_page)
            print(f"  📄 Page {page} → {collected}/{MAX_RESULTS} collected")
            page += 1

            # Be polite to the API - avoid rate limiting
            time.sleep(0.5)

            # If the API returned fewer results than requested, no more pages
            if len(jobs_page) < needed:
                break

        except Exception as e:
            print(f"  ⚠️ Exception on '{title}' page {page}: {e}")
            time.sleep(2)
            break

print(f"\n{'='*50}")
print(f"✅ Total jobs collected: {len(all_jobs)}")

# --- Save to CSV ---
df = pd.DataFrame(all_jobs)

# Reorder columns
df = df[["URL", "Position", "Company", "Location", "Job_Description", "Search_Title", "Date_Scraped"]]

# Drop exact duplicates (same URL)
before = len(df)
df.drop_duplicates(subset=["URL"], inplace=True)
print(f" Removed {before - len(df)} duplicate jobs")
print(f" Final unique jobs: {len(df)}")

filename = f"jooble_jobs_{datetime.today().strftime('%Y-%m-%d')}.csv"
df.to_csv(filename, index=False, encoding="utf-8-sig")  # utf-8-sig for Excel compatibility

print(f"\n💾 Saved to: {filename}")
print(df.head(3))


```
